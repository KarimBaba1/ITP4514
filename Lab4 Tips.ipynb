# Q2

```python
# N-Queen with CSP

from constraint import Problem

# Set the size of the chessboard (n x n)
n = 5

# Create a new constraint satisfaction problem instance
problem = Problem()

# Define the columns and rows (from 1 to n)
cols = range(1, n + 1)
rows = range(1, n + 1)

# Add variables to the problem. Each column can take a value from the rows.
# This means each column will have a queen placed in one of the rows.
problem.addVariables(cols, rows)

# Add constraints to ensure that no two queens can attack each other
# Iterate over pairs of columns
for col1 in cols:
    for col2 in cols:
        if col1 < col2:  # Only consider each pair once
            # Add a constraint for the current pair of columns
            problem.addConstraint(
                # Lambda function to check the conditions for the two queens
                lambda row1, row2, col1=col1, col2=col2:
                # Ensure the queens are not in the same row
                row1 != row2 and
                # Ensure the queens are not on the same diagonal
                abs(row1 - row2) != abs(col1 - col2),
                # Specify the variables (columns) to which this constraint applies
                (col1, col2)
            )

# Retrieve all possible solutions that satisfy the constraints
solutions = problem.getSolutions()

# Print each solution found
for sol in solutions:
    print(sol)  # Each solution is a dictionary mapping columns to rows

# Print the total number of solutions found
print('\nSolutions found: %i' % len(solutions))
```
