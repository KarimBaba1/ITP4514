```python
n = 5  # set any n (n>=3 has nontrivial solutions; n=1 trivial; n=2 none)

magic = n * (n*n + 1) // 2
Ldiag = [i*n + i for i in range(n)]
Rdiag = [i*n + (n - 1 - i) for i in range(n)]

problem = Problem()
problem.addVariables(range(n*n), range(1, n*n + 1))
problem.addConstraint(AllDifferentConstraint(), range(n*n))
problem.addConstraint(ExactSumConstraint(magic), Ldiag)
problem.addConstraint(ExactSumConstraint(magic), Rdiag)

for r in range(n):
    problem.addConstraint(ExactSumConstraint(magic), [r*n + c for c in range(n)])

for c in range(n):
    problem.addConstraint(ExactSumConstraint(magic), [c + n*r for r in range(n)])

solution = problem.getSolution()
print(solution)

# Optional pretty-print
if solution:
    for r in range(n):
        print([solution[r*n + c] for c in range(n)])
else:
    print("No solution found.")
```

```python
# !pip install unification  # run once in notebooks if needed

from unification import var            # var comes from unification
from unification.match import match    # match decorator comes from unification.match

n = var('n')  # pattern variable

# --- Fibonacci (same style) ---
@match(0)
def fib(n):
    return 0

@match(1)
def fib(n):
    return 1

@match(n)
def fib(n):
    return fib(n - 1) + fib(n - 2)

# --- Factorial (same style) ---
@match(0)
def factl(n):
    return 1  # 0! = 1

@match(n)
def factl(n):
    return n * factl(n - 1)


# quick sanity checks
print(f"fib(10)  = {fib(10)}")   # 55
print(f"factl(5) = {factl(5)}")  # 120
```

## Q4b

```python
def father(x, y):
    return lall(parent(x, y), male(x))
```

```python
def mother(x, y):
    return lall(parent(x, y), female(x))
```

```python
def brother(x, y):
    z = var()
    return lall(male(x), neq(x, y), parent(z, x), parent(z, y))
```

```python
def sister(x, y):
    z = var()
    return lall(female(x), neq(x, y), parent(z, x), parent(z, y))
```

```python
X = var()

print("Fathers of Amy:   ", run(0, X, father(X, "Amy")))
print("Mothers of Amy:   ", run(0, X, mother(X, "Amy")))
print("Brothers of Amy:  ", run(0, X, brother(X, "Amy")))
print("Sisters of Bobby: ", run(0, X, sister(X, "Bobby")))
```
