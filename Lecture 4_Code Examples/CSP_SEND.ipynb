## CSP : SEND + MORE = MONEY

```python
from constraint import *

letters = ['s','e','n','d','m','o','r','y']
digits = range(10)

def sum_constraint(s, e, n, d, m, o, r, y):
    send = s * 1000 + e * 100 + n * 10 + d
    more = m * 1000 + o * 100 + r * 10 + e
    money = m * 10000 + o * 1000 + n * 100 + e * 10 + y
    return send + more == money

problem = Problem()
for l in letters:
    problem.addVariable(l, digits)

problem.addConstraint(AllDifferentConstraint(), letters)
problem.addConstraint(sum_constraint, letters)
# problem.addConstraint(lambda x: x != 0, ['m'])

solutions = problem.getSolutions()
for s in solutions:
    print(s)
```
